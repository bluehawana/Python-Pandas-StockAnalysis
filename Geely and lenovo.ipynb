```python
#Define the data for analysis
gafataDict={"Geely":"0175.hk","Facebook":"FB","Apple":"AAPL","Alibaba":"BABA","Lenovo":"0092.hk"}
```

```python
start_date="2000-1-1"
end_date="2019-5-1"
```

```python
import pandas as pd
from pandas_datareader import data as pdr
import datetime
import yfinance as yf 
yf.pdr_override()
%config InlineBackend.figure_format = 'svg'
```

```python
Geely=pdr.get_data_yahoo(gafataDict["Geely"],start_date,end_date)
Lenovo=pdr.get_data_yahoo(gafataDict["Lenovo"],start_date,end_date)
```
