```python
#Define the data for analysis
gafataDict={"Geely":"0175.hk"}
```

```python
start_date="2005-1-1"
end_date="2020-5-3"
```

```python
import pandas as pd
from pandas_datareader import data as pdr
import datetime
import yfinance as yf 
yf.pdr_override()
%config InlineBackend.figure_format = 'svg'
```

```python
Geely=pdr.get_data_yahoo(gafataDict["Geely"],start_date,end_date)
```

```python
Geely.describe()
```

```python
import matplotlib.pyplot as plt  
plt.plot(Geely["Close"])  
plt.legend(['Geely'],loc='upper center') 
plt.ylabel('HK$')
plt.title("Geely stock performance before and after the M&A deal")  
plt.grid(True)  
plt.show()  
```
